```python
import pandas as pd  # Voor tabellen en data-analyse
import numpy as np  # Voor getallen en berekeningen
import matplotlib.pyplot as plt  # Voor grafieken

# Stap 2: Definieer de maanden van het onderzoek
maanden = ["Augustus 2023", "September 2023", "Oktober 2023", "November 2023", 
           "December 2023", "Januari 2024", "Februari 2024", "Maart 2024"]


aantal_huishoudens = 19

data = []
```

```python
# Stap 5: Willekeurige gegevens maken voor elk huishouden en elke maand
np.random.seed(42)  # Dit zorgt ervoor dat de data elke keer hetzelfde is

for maand in maanden:
    for huisnummer in range(1, aantal_huishoudens + 1):
        energie_verbruik = np.random.randint(280, 450)  # Energieverbruik per huishouden (kWh)
        zonne_energie = np.random.randint(100, 350)  # Hoeveel zonne-energie wordt opgewekt (kWh)
        batterij_opslag = np.random.randint(50, 100)  # Hoeveel energie wordt opgeslagen (kWh)
        netto_energie = energie_verbruik - zonne_energie  # Hoeveel energie overblijft na opwekking

        # Voeg deze data toe aan de lijst
        data.append([maand, huisnummer, energie_verbruik, zonne_energie, batterij_opslag, netto_energie])
```

```python
# Stap 6: Zet de data in een tabel (DataFrame)
df = pd.DataFrame(data, columns=["Maand", "Huishouden ID", "Energieverbruik (kWh)", 
                                 "Opgewekte zonne-energie (kWh)", "Batterij-opslagcapaciteit (kWh)", 
                                 "Netto energieverbruik (kWh)"])

# Stap 7: Opslaan als Excel-bestand
bestand_naam = "Energieverbruik_Huishoudens.xlsx"
df.to_excel(bestand_naam, index=False)

print(f"✅ De dataset is opgeslagen als: {bestand_naam}")
```

```python
# Gemiddelde waarden per maand berekenen
df_gemiddelde = df.groupby("Maand").mean(numeric_only=True)

# Grafiek 1: Energieverbruik vs. Zonne-energie
plt.figure(figsize=(10, 5))
plt.plot(df_gemiddelde.index, df_gemiddelde["Energieverbruik (kWh)"], label="Gemiddeld Energieverbruik", marker='o')
plt.plot(df_gemiddelde.index, df_gemiddelde["Opgewekte zonne-energie (kWh)"], label="Gemiddelde Zonne-energie", marker='s')
plt.xticks(rotation=45)
plt.ylabel("kWh per maand")
plt.title("Gemiddeld Energieverbruik vs. Opgewekte Zonne-energie")
plt.legend()
plt.grid()
plt.show()
```
